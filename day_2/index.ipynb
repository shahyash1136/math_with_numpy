## Day 2 — Vector Addition & Scalar Multiply

```python
import numpy as np
from typing import Any
```

### Element-wise operations

```python
v = np.array([1, 2])
x = np.array([3, -4])

addition = v + x
print("Addition", addition)

subtraction = v - x
print("Subtraction", subtraction)

multiplication = v * 3 # scaler multiplication
print("Multiplication", multiplication)

division = v / 2 # Scaler division
print("Division", division)
```

```python
v = np.array([1, 2, 3])
x = np.array([4, 5, 6])
addition = v + x
print("Addition", addition)
```

### Scaler common function

```python
def scale_and_add(v1: list[int | float], v2: list[int | float], scaler: int | float):
    if not v1 or not v2:
        raise ValueError("List cannot be empty")

    if not scaler:
        raise ValueError("Scaler cannot be blank")

    addition = np.array(v1) + np.array(v2)
    return addition * scaler


data1 = scale_and_add([1, 2], [4, 5], 2)
data2 = scale_and_add([-1, -5, 5], [4, 5, 6], 3)
data3 = scale_and_add([3, 1, 2, 6], [2, 4, 6, 9], 6)

print("\nData 1:", data1)
print("\nData 2:", data2)
print("\nData 3:", data3)
```
