## Day 2 — Vector Addition & Scalar Multiply

```python
import numpy as np
from typing import Union
```

### Element-wise operations

```python
v = np.array([1, 2])
x = np.array([3, -4])

addition = v + x
print("Addition", addition)

subtraction = v - x
print("Subtraction", subtraction)

multiplication = v * 3 # scaler multiplication
print("Multiplication", multiplication)

division = v / 2 # Scaler division
print("Division", division)
```

```python
v = np.array([1, 2, 3])
x = np.array([4, 5, 6])
addition = v + x
print("Addition", addition)
```

### Scaler common function

```python
def scale_and_add(v1: list, v2: list, scalar: Union[int, float]) -> np.ndarray:
    if not v1 or not v2:
        raise ValueError("Vectors cannot be empty")
    if len(v1) != len(v2):
        raise ValueError(
            f"Vectors must be the same length, got {len(v1)} and {len(v2)}"
        )
    if scalar is None:
        raise ValueError("Scalar cannot be None")

    return scalar * np.array(v1) + np.array(v2)


# Test 1 — basic positive values
print(scale_and_add([1, 2], [4, 5], 2))  # expected: [6, 9]

# Test 2 — negative values in v1
print(scale_and_add([-1, -5, 5], [4, 5, 6], 3))  # expected: [1, -10, 21]

# Test 3 — scalar = 0 (v1 gets zeroed out)
print(scale_and_add([3, 1, 2, 6], [2, 4, 6, 9], 0))  # expected: [2, 4, 6, 9]

# Test 4 — float scalar
print(scale_and_add([1, 2], [0, 0], 0.5))  # expected: [0.5, 1.0]
```
