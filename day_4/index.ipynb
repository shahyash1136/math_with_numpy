## Day 4 — Norms & Distance 

```python
import numpy as np
```

#### What is Norm
- A norm measures the size of length of a vector or matrix

```python
vector = np.array([3,4])
norm = np.linalg.norm(vector)
print(norm)
```

### Matrix norm example

```python
matrix = np.array([[1, 2], [3, 4]])
norm = np.linalg.norm(matrix)

print(norm)
```

```python
vector = np.array([3, 4])
l1_norm = np.linalg.norm(
    vector, ord=1
)  # Total walking distance (Sum of absolute values)
l2_norm = np.linalg.norm(vector, ord=2)  # Straight line distance (Euclidean distance)
inf_norm = np.linalg.norm(vector, ord=np.inf)  # Biggest step (Max absolute value)

print(l1_norm)
print(l2_norm)
print(inf_norm)
```

```python
vector_a = np.array([1, 2])
vector_b = np.array([4, 6])

norm = np.linalg.norm(vector_a - vector_b) # Euclidean Distance
print(norm)
```

```python
vector_a = np.array([1, 2])
vector_b = np.array([4, 6])
vector_c = np.array([0, 0])

distances = {
    "AB": np.linalg.norm(vector_a - vector_b),
    "BC": np.linalg.norm(vector_b - vector_c),
    "CA": np.linalg.norm(vector_c - vector_a),
}

# sort by distance
sorted_distances = sorted(distances.items(), key=lambda x: x[1])

for pair, dist in sorted_distances:
    print(f"{pair}: {dist:.2f}")
```
