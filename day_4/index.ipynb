## Day 4 — Norms & Distance 

```python
import numpy as np
```

#### What is Norm
- A norm measures the size of length of a vector or matrix

```python
vector = np.array([3,4])
norm = np.linalg.norm(vector)
print(norm)
```

### Matrix norm example

```python
matrix = np.array([[1, 2], [3, 4]])
norm = np.linalg.norm(matrix)

print(norm)
```

```python
vector = np.array([3, 4])
l1_norm = np.linalg.norm(vector, ord=1)
l2_norm = np.linalg.norm(vector, ord=2)
inf_norm = np.linalg.norm(vector, ord=np.inf)

print(l1_norm)
print(l2_norm)
print(inf_norm)

```
