## Day 7 — Matrix Multiplication

```python
import numpy as np
```

```python
matrix_A = np.array([[1, 2], [3, 4]])
matrix_B = np.array([[5, 6], [7, 8]])

# 1. np.matmul(A, B) and A @ B — same result for 2D arrays
print(np.array_equal(np.matmul(matrix_A, matrix_B), matrix_A @ matrix_B))

# 2. Verify your hand calc: result should be [[19,22],[43,50]]
expected = np.array([[19, 22], [43, 50]])
print(np.array_equal(matrix_A @ matrix_B, expected))

# 3. Show it's NOT commutative: A @ B ≠ B @ A
print(np.array_equal(matrix_A @ matrix_B, matrix_B @ matrix_A))
```

```python
# ML demo: X = np.random.rand(10,3), W = np.random.rand(3,2) → X @ W gives (10,2)

X = np.random.rand(10, 3)
W = np.random.rand(3, 2)
multiplication = X @ W
print(multiplication.shape)
print(multiplication)
```
