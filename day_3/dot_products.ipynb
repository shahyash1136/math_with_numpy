```python
import numpy as np
```

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

dot_product = np.dot(a,b)
print(dot_product)
```

```python
print(np.dot(np.array([1, 0]), np.array([0, 1])))
```

```python
def is_perpendicular(a: list, b: list) -> bool:
    if not a or not b:
        raise ValueError("Vectors cannot be empty")
    if len(a) != len(b):
        raise ValueError(f"Vectors must be the same length, got {len(a)} and {len(b)}")

    return bool(np.isclose(np.dot(np.array(a), np.array(b)), 0))


# Test 1
print(is_perpendicular([1, 0], [0, 1]))  # True ✅

# Test 2
print(is_perpendicular([2, 3], [5, 6]))  # False ✅

# Test 3
print(is_perpendicular([5, 5], [6, -6]))  # True ✅

# Test 4
a = [1 / np.sqrt(2), 1 / np.sqrt(2)]
b = [-1 / np.sqrt(2), 1 / np.sqrt(2)]
print(is_perpendicular(a, b))
```
